```python
%%R
## Create a vector
x <- rep(c("DS","AI","IoT","MI"), c(8,10,5,2))
print(x)
#Count number of occurances of each character string
mytable <- table(x)
print(mytable)
# Bar plot using that table
barplot(mytable)
```

```python
%%R
barplot(mytable,col=5:8,main="bar plot",las=0)
```

```python
%%R
# Create a vector 
x <- rep(c("IoT", "ML", "DS", "AI"), c(8,10,5,2))

# Count number of occurences of each string.
mytable <- table(x)

pie(mytable,
    main="The Pie Chart",
    col=c("lightblue", "lightgreen", "salmon", "black"))
```

```python
%%R
# Create a matrix of 1000 random values from the normal distribution
x <- matrix(rnorm(1000), ncol=4)
xdf=as.data.frame(x)
print(xdf)
# Basic boxplot
boxplot(xdf) 
```

```python
%%R
# Vector of 200 random values from normal distribution
x <- rnorm(200)

# Plot histogram
hist(x)
print(mean(x))
print(sd(x))
```

```python
%%R
hist(x,
     border="blue",
     breaks=50,
     main="Histrogram",
     xlab="",
     cex.main=2.5,
     cex.lab=2)
```

```python
%%R
library(ggplot2)
```

```python
%%R
data("iris")
ggplot(data=iris, mapping=aes(x=Species, y=Sepal.Length))
```

```python
%%R
data("iris")
ggplot(data=iris, mapping=aes(x=Species, y=Sepal.Length))+ geom_point()
```

```python
%%R
ggplot(data=iris, mapping=aes(x=Sepal.Width, y=Sepal.Length))+ geom_point()
```

```python
%%R
ggplot(data=iris, mapping=aes(x=Sepal.Width, y=Sepal.Length))+ geom_point(color="blue")
```

```python
%%R
ggplot(data=iris, mapping=aes(x=Sepal.Width, y=Sepal.Length,color=Species))+ 
geom_point()
```

```python
%%R
ggplot(data=iris, mapping=aes(x=Sepal.Width, y=Sepal.Length,fill=Species))+ 
geom_boxplot()
```

```python
%%R
ggplot(data=iris, mapping=aes(x=Sepal.Length,fill=Species))+ 
geom_bar()
```

```python
%%R
ggplot(data=iris, mapping=aes(x=Sepal.Length,fill=Species))+ 
geom_histogram()
```

```python
%%R
ggplot(data=iris, mapping=aes(x=Sepal.Length,fill=Species))+ 
geom_histogram(position="dodge")
```
